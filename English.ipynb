```python
f = open("GG.txt", "r")

names = []

while True:

    line = f.readline()

    if line == line.upper():

      line = line.replace('\n', '')
      line = line.replace(' ', '')

      ff = open(f'{line}.txt', 'w')

      names.append(line)

      f.readline()
      f.readline()

    elif line == "":

      ff.close()

    else:

      ff.write(line)

    if not line:

      break

    f.readline()

f.close()

names.remove('')
```

```python
import pprint

m = []

for s in names:

  fk = []

  f = open(f'{s}.txt', 'r')

  while True:

    eng = f.readline().replace('\n', '')

    if eng == '':

      f.close()

      break

    rus = f.readline().replace('\n', '')

    fk.append([eng, rus])

  m.append(fk)
```
